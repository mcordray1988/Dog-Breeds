```python
browser = Browser("firefox")
```

```python
breeder = []
availability = []
location = []
blood = []
award = []
```

```python
for i in range(900):
    url = f'https://marketplace.akc.org/puppies?page={i+1}'
    browser.visit(url)
    
    html = browser.html
# Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
# Retrieve all elements that contain breeder information
    
    for article in soup.find_all('a', class_='sc-kgAjT kFFrdz'):
        name = article.h4.text.strip()
        description = article.find('div', class_='breeder-litter-card__details').p.text.strip()
        loc = article.find('p', class_='breeder-card__location').text.strip()
        if(i < 87):
            badge = article.find('img').get('src')
        
            
        bloodline = article.find('div', class_='breeder-litter-card__champ').get('style')
        
        
        if bloodline == 'visibility: visible;':
            new_bloodline = "True"
        else:
            new_bloodline = "False"

        if badge == 'https://cdn.marketplace.akc.org/site-media/bom-icons/platinum.sm.png':
            new_badge = 'platinum'
        elif badge == 'https://cdn.marketplace.akc.org/site-media/bom-icons/gold.sm.png':
            new_badge = 'gold'
        elif badge == 'https://cdn.marketplace.akc.org/site-media/bom-icons/silver.sm.png':
            new_badge = 'silver'
        elif badge == 'https://cdn.marketplace.akc.org/site-media/bom-icons/bronze.sm.png':
            new_badge = 'bronze'
        else:
            new_badge = 'standard'

            breeder.append(name)
            availability.append(description)
            location.append(loc)
            award.append(new_badge)
            blood.append(new_bloodline)


    #print(name)
    
    #print(new_badge)
    #print(description)
    #print(loc)
    #print(new_bloodline)
    #print("-------------------------")

    try:
        browser.links.find_by_partial_text('»').click()
        time.sleep(1)

    except Exception:
        pass
```
