```python
browser = Browser("firefox")
```

```python
breeder = []
availability = []
location = []
```

```python
url = f'https://marketplace.akc.org/puppies?page=1'
browser.visit(url)
    
html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')
# Retrieve all elements that contain breeder information
breeding = soup.find('div', class_='sc-dVhcbM gdJZAt')
articles = breeding.find('a', class_='sc-kgAjT kFFrdz')
print(breeding.text.strip())

# Iterate through each breeder
# for breed in breeding:
for x in breeding:

    soupy = articles.find('div', class_='breeder-litter-card__details')
    div = soupy.find('div')
    
    h4 = div.find('h4', class_='breeder-card__name')

    name = h4.text.strip()
    breeder.append(name)
    p = div.find('p')
    description = p.text.strip()
    p2 = div.find('p', class_='breeder-card__location')
    loc = p2.text.strip()
    

    print(name)
    print(description)
    print(loc)
    print("-------------------------")

```
